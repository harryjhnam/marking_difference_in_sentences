## Path to two input sentences

```python
sen1_path = './test/sentence.txt'
sen2_path = './test/wrong_sentence.txt'
```

### Read lines from input paths

```python
with open(sen1_path, 'r') as f:
    sen1 = f.readlines()

with open(sen2_path, 'r') as f:
    sen2 = f.readlines()
```

### Import and set nltk library

```python
import nltk
nltk.download('punkt')
```

### Function comparing words in sentences

```python
import difflib
import sys

def compare_tokenized_sentences(token_sen1, token_sen2):
    diff = difflib.Differ()
    result = list(diff.compare(token_sen1, token_sen2))
    return result
    
```

### Tokenizing words

```python
from nltk.tokenize import word_tokenize

for i, line1 in enumerate(sen1):
    tokens1 = word_tokenize(line1)
    tokens2 = word_tokenize(sen2[i])
    print(compare_tokenized_sentences(tokens1, tokens2))
```

## Path to two input sentences

```python
sen1_path = '../speech2text/dataset/final_dataset_5/split_train.vi'
sen2_path = '../speech2text/dataset/final_dataset_5/split_train.en'
```

### Read lines from input paths

```python
with open(sen1_path, 'r') as f:
    sen1 = f.readlines()

with open(sen2_path, 'r') as f:
    sen2 = f.readlines()
```

### Import and set nltk library

```python
import nltk
nltk.download('punkt')
```

### Function comparing words in sentences

```python
import difflib
import sys

def compare_tokenized_sentences(token_sen1, token_sen2):
    diff = difflib.Differ()
    result = list(diff.compare(token_sen1, token_sen2))
    return result
    
```

### Tokenizing words

```python
from nltk.tokenize import word_tokenize

compare_results = []

num_samples = len(sen1)
samples_per_1percent = num_samples//100

for i, line1 in enumerate(sen1):
    
    line2 = sen2[i]
    
    tokens1 = word_tokenize(line1)
    tokens2 = word_tokenize(line2)
    
    result = compare_tokenized_sentences(tokens1, tokens2)
    
    compare_results.append(result)
    '''
    if i%samples_per_1percent == 0:
        print("{}% done...".format(i//num_samples))
        print("Sample : ")
        print("   INPUT 1 : ", line1)
        print("   INPUT 2 : ", line2)
        print("   OUTPUT  : ", result)
        print("")
        print("")
    '''
        
print("")
print("DONE")
```

```python
def is_prediction_correct_labeling(compare_result_line):
    
    is_repeated = False
    prev_token = " "
    
    return_label = []
    
    for word in compare_result_line:
                
        if word[0] == " ":
            return_label.append(0)
        else:
            if word[0] == prev_token:
                is_repeated = True
                return_label.append(1)
            else:
                is_repeated = False
                return_label.append(0)
            
        prev_token = word[0]
        
    return return_label
            
```
